```python
import os
from utils import imageSynthesiser
import matplotlib.pyplot as plt
import numpy as np

join = os.path.join
rootDir = os.path.dirname(os.getcwd()) # project parent directory
```

```python
# Initialise the image synthesiser. Note that a background image and lesion mask are randomly
# sampled each time this is run! In the following example a new traiing example with macular 
# atrophy is synthesised by extracting the lesion from another MMAC image (with macular atrophy).
synthesiser = imageSynthesiser(dataDir = join(rootDir, "data"), 
                               lesion = 'MA', 
                               lesionDataset = 'MMAC', 
                               lesionBankDir = join(rootDir, "lesionBank"))
```

```python
# BI  = background image (without macular atrophy)
# LI  = lesion image (with macular atrophy)
# syn = synthesised image (new example with macular atrophy)
BI, LI, syn = synthesiser.synthesise(x_offset = 0,
                                     y_offset = 0,
                                     rotateAngle = -25,
                                     blurFilterSize = 20)

# From left to right: background image (MMAC), image from which 
# lesion is extracted (MMAC or PALM), synthesised image.
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,10))
ax1.imshow(np.array(BI)); ax1.axis('off')
ax2.imshow(np.array(LI)); ax2.axis('off')
ax3.imshow(np.array(syn)); ax3.axis('off')
plt.show()
```
